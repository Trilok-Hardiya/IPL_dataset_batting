```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/IPL_player_performance_analysis/Batting_data_cleaned/Batting_data_cleaned.csv')
```

```python
selected_columns = df[['runs', 'strikeRate', 'sixes','fours']]
sns.pairplot(selected_columns)
plt.title('Pair Plot of Player Statistics')
plt.show()
```

```python
 run_count=df.groupby(['season'])['runs'].sum()
print(run_count)
```

```python
# Plotting a donut chart
plt.figure(figsize=(10, 7))
plt.pie(run_count, labels=run_count, startangle=140, autopct='%1.1f%%')
centre_circle = plt.Circle((0, 0), 0.70, fc='white')  # Create a white circle at the center
fig = plt.gcf()
fig.gca().add_artist(centre_circle)  # Add circle to the pie chart

# Adding title
plt.title('Total Runs Scored in Each IPL Season')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is a circle.
plt.show()
```

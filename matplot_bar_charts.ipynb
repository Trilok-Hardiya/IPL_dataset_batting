```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file_path='/IPL_player_performance_analysis/Batting_data_cleaned/Batting_data_cleaned.csv'
df=pd.read_csv(file_path)
# run_count=df.loc[df.groupby(['season','name'])['runs'].sum()].nlargest(5,'runs')

# print(run_count)
top_5_scorers_per_season = df.groupby('season').apply(lambda x: x.nlargest(3, 'runs')).reset_index(drop=True)

# Plotting
plt.figure(figsize=(20, 8))
for season in top_5_scorers_per_season['season'].unique():
    season_data =top_5_scorers_per_season[top_5_scorers_per_season['season'] == season]
    plt.bar(season_data['name'], season_data['runs'], label=f"Season {season}")

plt.xlabel('Players')
plt.ylabel('Total Runs')
plt.title('Highest Run Scorers in a single match for each Season')
plt.xticks(rotation=60)
plt.legend()
plt.show()
```

```python

# Plotting the top run scorers
plt.figure(figsize=(15, 6))
plt.bar(df['season'], df['runs'], color='orange')

# Adding labels and title
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.title('Top Run Scored in Each IPL Season')


# Show plot
plt.show()
```

```python
# Group by 'season' and 'name', summing the runs, then resetting the index
total_runs = df.groupby(['season', 'name'])['runs'].sum().reset_index()

# Get the top 5 scorers per season
top_5_scorers = total_runs.groupby('season').apply(lambda x: x.nlargest(3, 'runs')).reset_index(drop=True)
```

```python

# Plotting the top run scorers
plt.figure(figsize=(6, 15))
top_5_scorers['season']=top_5_scorers['name']+' '+top_5_scorers['season'].astype(str)
plt.barh(top_5_scorers['season'], top_5_scorers['runs'], color='red')
# Adding labels and title
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.title('Top Run Scored in Each IPL Season')


# Show plot
plt.show()
```
